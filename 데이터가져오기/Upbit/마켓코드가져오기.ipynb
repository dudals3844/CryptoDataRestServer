```python
import requests
import pandas as pd
import json
```

```python
url = 'https://api.upbit.com/v1/market/all'
```

```python
response = requests.get(url)
```

```python
data_dict = json.loads(response.text)
```

```python
pd.DataFrame(data_dict)
```

```python
insert_url = 'http://127.0.0.1:8000/upbit/ticker/'
```

```python
for data in data_dict:
    response = requests.post(insert_url, data=data)
    print(response.text)
```
